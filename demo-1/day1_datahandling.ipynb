```python
data = {
    "Salary" : [25000, 30000, 37000, 47000, 28000, 39000, 55000, 52000, 31000, 60000, 80000, 72000, 23000, 22000, 66000, 65000, 34000, 32000, 59000, 45000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
plt.plot(df["Salary"], color = "red", marker = "o", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
#histogram

plt.hist(df["Salary"], bins = 5, color="green")
plt.show()
```

```python
# boxplot

plt.boxplot(df["Salary"])
plt.show()
```

```python
plt.boxplot(df["Salary"])
plt.show()
```

```python
df["dept"] = ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'] * 2
```

```python
df.head()
```

```python
df.info()
```

```python
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels = count.index, autopct = "%1.1f", explode = [0, 0.2, 0.1])
plt.show()
```

```python
# Countplot

plt.bar(count.index, count, color = ["green", "red", "blue"])
plt.show()
```

```python
df["Age"] = [21, 24, 31, 22, 35, 45, 38, 22, 43, 32, 36, 26, 25, 44, 33, 27, 37, 42, 39, 29]
df.head()
```

```python
plt.scatter(df["Age"], df["Salary"], color="orange")
plt.show()
```

```python
sort_age = df.sort_values("Age")
```

```python
plt.plot(sort_age["Age"], df["Salary"], color = "red", marker = "*", linestyle=":")
plt.grid()
plt.show()
```

```python
plt.bar(sort_age["Age"], df["Salary"], color = "green")
plt.show()
```

```python
hr_sal = df[df["dept"] == "HR"]["Salary"]
hr_sal
```

```python
fin_sal = df[df["dept"] == "Finance"]["Salary"]
fin_sal
```

```python
it_sal = df[df["dept"] == "IT"]["Salary"]
it_sal
```

```python
plt.boxplot([hr_sal, fin_sal, it_sal], labels = ["HR", "Finance", "IT"])
plt.show()
```

```python
sal_by_dept = df.groupby("dept")["Salary"].sum()
sal_by_dept
```

```python
plt.pie(sal_by_dept, labels = sal_by_dept.index, autopct = "%1.2f")
plt.axis("equal")
plt.show()
```
